```python
import mysql.connector
from mysql.connector import Error

def connect_to_database():
    try:
        connection = mysql.connector.connect(
                host=db_host,
                database=db_name,
                user=db_user,
                password=db_password
            )
        
        if connection.is_connected():
            cursor = connection.cursor()
            cursor.execute("SHOW TABLES")
            tables = cursor.fetchall()
            
            print("Nombres de las tablas en la base de datos:")
            for table in tables:
                print(table[0])
                
            cursor.close()
            connection.close()
            print("Conexión a MySQL cerrada")
    
    except Error as e:
        print(f"Error al conectar a MySQL: {e}")

if __name__ == "__main__":
    connect_to_database()
```

```python
import pandas as pd
import mysql.connector



# Conexión a la base de datos
conn = mysql.connector.connect(
    host=db_host,
    database=db_name,
    user=db_user,
    password=db_password
)

# Cursor para ejecutar consultas
cursor = conn.cursor()

# Consulta para obtener información de las tablas y columnas
query = """
    SELECT 
        TABLE_NAME, 
        COLUMN_NAME, 
        DATA_TYPE
    FROM 
        INFORMATION_SCHEMA.COLUMNS 
    WHERE 
        TABLE_SCHEMA = %s
    ORDER BY TABLE_NAME
"""

# Ejecutar la consulta y obtener los resultados
cursor.execute(query, (db_name,))
results = cursor.fetchall()

# Inicializar variables para mantener el nombre de la tabla actual
current_table = None

# Recorrer los resultados y mostrar la información en el formato deseado
for result in results:
    table_name = result[0]
    column_name = result[1]
    data_type = result[2]
    
    # Si cambiamos de tabla, mostrar el nombre de la nueva tabla
    if table_name != current_table:
        print(f"\n{table_name}\n")
        current_table = table_name
    
    # Mostrar el nombre de la columna y su tipo de dato
    print(f"{column_name} {data_type}")

# Cerrar el cursor y la conexión a la base de datos
cursor.close()
conn.close()
```

```python
import pandas as pd
import numpy as np

def crear_df_pavisivo(dataframe):
    
    pasivosCaja   = np.random.choice(cajas_percepcion_pasiva)
    pasivosCausa = np.random.choice(causas_percepciones_pasivas)
    pasivosEstado = np.random.choice(estados_percepcion_pasiva)
    pasivosRegimen = np.random.choice(regimenes_percepcion_pasiva)
    
    new_data = pd.DataFrame([{'pasivosCaja' :pasivosCaja ,'pasivosCausa': pasivosCausa, 'pasivosEstado':pasivosEstado,  'pasivosRegimen':pasivosRegimen}])
    dataframe = pd.concat([dataframe, new_data], ignore_index=True)
    
    return dataframe
```

```python
import mysql.connector

def obtener_dnies(db_host, db_user, db_password, db_name):
    # Crear una conexión a la base de datos MySQL
    conn = mysql.connector.connect(
        host=db_host,
        user=db_user,
        password=db_password,
        database=db_name
    )

    # Crear un cursor
    cursor = conn.cursor()

    # Ejecutar la consulta
    cursor.execute("SELECT docenteDni FROM docentes")

    # Obtener los resultados
    resultado = cursor.fetchall()

    # Crear una lista de DNIs
    lista_dnis = [row[0] for row in resultado]

    # Cerrar el cursor y la conexión
    cursor.close()
    conn.close()

    # Devolver la lista de DNIs
    return lista_dnis

# Llamar a la función y mostrar el resultado


lista_dnis = obtener_dnies(db_host, db_user, db_password, db_name)
print(lista_dnis)

```

```python
import random
import datetime

def fecha_aleatoria():
    # Generar un año aleatorio entre 1900 y 2023
    year = random.randint(1900, 2023)
    
    # Generar un mes aleatorio entre 1 y 12
    month = random.randint(1, 12)
    
    # Generar un día aleatorio dependiendo del mes y si es un año bisiesto o no
    if month in [1, 3, 5, 7, 8, 10, 12]:
        day = random.randint(1, 31)
    elif month in [4, 6, 9, 11]:
        day = random.randint(1, 30)
    else:
        if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
            day = random.randint(1, 29)  # Año bisiesto
        else:
            day = random.randint(1, 28)  # Año no bisiesto
    
    # Crear y devolver el objeto de fecha
    fecha = datetime.date(year, month, day)
    return fecha


```
