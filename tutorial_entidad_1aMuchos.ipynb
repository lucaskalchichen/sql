```python
import pandas as pd
import mysql.connector



# Conexión a la base de datos
conn = mysql.connector.connect(
    host=db_host,
    database=db_name,
    user=db_user,
    password=db_password
)

# Cursor para ejecutar consultas
cursor = conn.cursor()

# Consulta para obtener información de las tablas y columnas
query = """
    SELECT 
        TABLE_NAME, 
        COLUMN_NAME, 
        DATA_TYPE
    FROM 
        INFORMATION_SCHEMA.COLUMNS 
    WHERE 
        TABLE_SCHEMA = %s
    ORDER BY TABLE_NAME
"""

# Ejecutar la consulta y obtener los resultados
cursor.execute(query, (db_name,))
results = cursor.fetchall()

# Inicializar variables para mantener el nombre de la tabla actual
current_table = None

# Recorrer los resultados y mostrar la información en el formato deseado
for result in results:
    table_name = result[0]
    column_name = result[1]
    data_type = result[2]
    
    # Si cambiamos de tabla, mostrar el nombre de la nueva tabla
    if table_name != current_table:
        print(f"\n{table_name}\n")
        current_table = table_name
    
    # Mostrar el nombre de la columna y su tipo de dato
    print(f"{column_name} {data_type}")

# Cerrar el cursor y la conexión a la base de datos
cursor.close()
conn.close()
```

```python
import mysql.connector

def contactosDni(db_host, db_user, db_password, db_name):
    # Crear una conexión a la base de datos MySQL
    conn = mysql.connector.connect(
        host=db_host,
        user=db_user,
        password=db_password,
        database=db_name
    )

    # Crear un cursor
    cursor = conn.cursor()

    # Ejecutar la consulta
    cursor.execute("SELECT docenteDni, docenteNombre, docenteApellido FROM docentes")

    # Obtener los resultados
    resultado = cursor.fetchall()

    # Crear el diccionario de resultados
    contactos = {}
    for row in resultado:
        dni, nombre, apellido = row
        contactos[dni] = [nombre, apellido]
        
    # Cerrar el cursor y la conexión
    cursor.close()
    conn.close()

    # Devolver el diccionario de contactos
    return contactos

# Llamar a la función y mostrar el resultado


contactos = contactosDni(db_host, db_user, db_password, db_name)
lista_dni = list(contactos.keys())

print(contactos, lista_dni)
```
