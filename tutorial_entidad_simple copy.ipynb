```python
import pandas as pd
import mysql.connector



# Conexión a la base de datos
conn = mysql.connector.connect(
    host=db_host,
    database=db_name,
    user=db_user,
    password=db_password
)

# Cursor para ejecutar consultas
cursor = conn.cursor()

# Consulta para obtener información de las tablas y columnas
query = """
    SELECT 
        TABLE_NAME, 
        COLUMN_NAME, 
        DATA_TYPE
    FROM 
        INFORMATION_SCHEMA.COLUMNS 
    WHERE 
        TABLE_SCHEMA = %s
    ORDER BY TABLE_NAME
"""

# Ejecutar la consulta y obtener los resultados
cursor.execute(query, (db_name,))
results = cursor.fetchall()

# Inicializar variables para mantener el nombre de la tabla actual
current_table = None

# Recorrer los resultados y mostrar la información en el formato deseado
for result in results:
    table_name = result[0]
    column_name = result[1]
    data_type = result[2]
    
    # Si cambiamos de tabla, mostrar el nombre de la nueva tabla
    if table_name != current_table:
        print(f"\n{table_name}\n")
        current_table = table_name
    
    # Mostrar el nombre de la columna y su tipo de dato
    print(f"{column_name} {data_type}")

# Cerrar el cursor y la conexión a la base de datos
cursor.close()
conn.close()
```

```python
import pandas as pd
import numpy as np

def crear_df_docentes(dataframe):
    while True:
        dni = np.random.randint(12000000, 39000000)
        if dni not in dataframe['docenteDni'].values:
            break
    
    sexo = np.random.choice(sexos)
    estado_civil = np.random.choice(estados_civiles)
    nacionalidad = np.random.choice(nacionalidades)
    
    if sexo == 'Masculino':
        nombre = np.random.choice(nombres_hombres)
    else:
        nombre = np.random.choice(nombres_mujeres)
    
    apellido = np.random.choice(apellidos)
    docenteCuit = generar_cuil(dni)
    docenteFecha = generar_fecha_nacimiento(estimar_fecha_nacimiento(dni))
    
    new_data = pd.DataFrame([{'docenteDni': dni, 'docenteSexo': sexo, 'docenteEstado_civil': estado_civil, 'docenteNacionalidad': nacionalidad, 
                             'docenteNombre': nombre, 'docenteApellido': apellido, 'docenteCuit': docenteCuit, 'docenteFecha_nac': docenteFecha}])
    dataframe = pd.concat([dataframe, new_data], ignore_index=True)
    
    return dataframe
```
